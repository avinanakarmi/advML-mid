```python
import matplotlib.pyplot as plt

methods = ['PCA + kMeans', 'cPCA + kMeans', 'kMeans']
ari_scores = [0.829, 0.562, 0.026]

plt.figure(figsize=(8, 6))
plt.bar(methods, ari_scores, color=['blue', 'green', 'purple'], alpha=0.2)

plt.xlabel('Methods')
plt.ylabel('Adjusted Rand Index (ARI)')
plt.title('Comparing ARI')

# Display the plot
plt.show()
```
